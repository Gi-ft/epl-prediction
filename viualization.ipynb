```python
# Distributin of final points
import matplotlib.pyplot as plt
import pickle
with open("sim_results.pkl", "rb") as f:
    sim_data = pickle.load(f)

arsenal_points_dist = sim_data["arsenal_points"]
city_points_dist = sim_data["city_points"]
point_diffs = sim_data["point_diff"]

plt.hist(arsenal_points_dist, bins=20, alpha=0.6, label='Arsenal')
plt.hist(city_points_dist, bins=20, alpha=0.6, label='Man City')

plt.legend()
plt.title("Distribution of Final Points")
plt.xlabel("Points")
plt.ylabel("Frequency")

plt.show()
```

```python
# Point Gap Analysis
plt.hist(point_diffs, bins=20)
plt.title("Point Difference (Arsenal - City)")
plt.xlabel("Point Gap")
plt.ylabel("Frequency")
plt.show()
```
